```python
import matplotlib.pyplot as plt
 
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
dfcomp = df[df['Occupation Title'].str.contains("Computer")]
```

```python
dfcomp= dfcomp.groupby(['Typical entry-level education']).count()[ 'Occupation Title']
dfcomp = pd.DataFrame(dfcomp)

dfcomp = dfcomp.reset_index()

labels = dfcomp['Typical entry-level education'].tolist()
sizes = dfcomp['Occupation Title'].tolist()
 
# Plot
plt.pie(sizes, labels=labels,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
matplotlib.rcParams.update({'font.size': 30})

fig = plt.gcf()
fig.set_size_inches(20,20) 

plt.title('Computer Professional by Degree')

plt.show()
```
